```python
import os
prob = []
level = []
types = []
solution = []
for filename in os.listdir('/content/drive/MyDrive/MWP-Classification/algebra'):
  f = open('/content/drive/MyDrive/MWP-Classification/algebra/'+filename)
  data = json.load(f)
  prob.append(data['problem'])
  level.append(data['level'])
  types.append(data['type'])
  solution.append(data['solution'])
  
```

```python
len(prob)
```

```python
import pandas as pd

df_algebra = pd.DataFrame({'Problem ': prob,"Level ":level,"Type":types,"Solution":solution})
```
