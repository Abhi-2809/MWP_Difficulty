```python
import pandas as pd
import csv
import re

```

```python
url = 'https://raw.githubusercontent.com/Abhi-2809/MWP_Difficulty/main/Algebra_train_cleaned.csv?token=AKC55RODPCX2KNNPPXBS4UDBRPCRM'
df = pd.read_csv(url)
```

```python
display(df)
```

```python
questions_words_only = []
questions_values_only = []
for row in df.iterrows():
  question = str(row[1][1])
  #print(question)
  extracted_words = re.findall(r'[a-z]+', question , flags = re.IGNORECASE)
  extracted_values = re.findall(r'[0-9]+', question , flags = re.IGNORECASE)
  questions_words_only.append(extracted_words)
  questions_values_only.append(extracted_values)
```

```python
print(str(questions_words_only))
print(str(questions_values_only))
```

```python
df['Extracted Words'] = questions_words_only
df['Extracted Values'] = questions_values_only
display(df)
```
